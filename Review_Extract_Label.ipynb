```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('C:\\Users\\Ammick\\Google Drive\\PHD\\Research Paper\\PYTHON\\2016 Movies Reviews\\All Data + Extraction Code\\Paper II Working\\Corrected_Bollywood_64_2016_reviews.xlsx')
```

```python
df=data.copy()
df=df[df.Ratings != '-']
df.describe()
```

```python
df['Ratings']=df['Ratings'].apply(lambda x: int(x))
```

```python
def Polarity(x):
    if x<4:
        return 'Very Negative'
    elif x>=4 and x<=5:
        return 'Negative'
    elif x>5 and x<7:
        return 'Neutral'
    elif x>=7 and x<=8:
        return 'Postive'
    else:
        return 'Very Postive'
    
df['Polarity']=df['Ratings'].apply(lambda x : Polarity(x))
df.hist()
df.plot()
plt.show()
```

```python
reviews_labeled=df[['Reviews','Polarity']]
reviews_labeled.to_excel('Review__Extract_Label.xlsx')
```
